## 2.0 Data Understanding

### 2.1 Dataset

I will be using Telcom's Customer Churn dataset (Churn_in_Telecom_Dataset.csv)

### 2.2 Columns Used 


- state: This indicates the state where the customer resides. 

- account length: This represents the number of months (or years) the customer has been with SyriaTel. 

- area code: This indicates the area code associated with the customer's phone number. 

- phone number: While not directly relevant for churn prediction itself (due to privacy concerns), this can be used for data validation or merging with other datasets containing customer information.

- international plan: Indicates whether the customer has subscribed to an international calling plan. 

- voice mail plan: Indicates whether the customer has subscribed to a voicemail plan. 

- number vmail messages: This shows the number of voicemail messages retrieved by the customer in the timeframe covered by the data. 

- total day minutes, total day calls, total day charge: These represent the total minutes spent on calls, number of calls made, and total charges incurred during daytime hours.

- total eve minutes, total eve calls, total eve charge: Similar to total day metrics, but for evening hours.

- total night minutes, total night calls, total night charge: Similar to total day metrics, but for nighttime hours.

- total intl minutes, total intl calls, total intl charge: These represent the total minutes spent on international calls, number of international calls made, and total charges incurred.

- customer service calls: This shows the number of times the customer contacted customer service in the timeframe covered by the data.

- churn: This is the binary target variable indicating whether the customer churned (1) or not (0) within the timeframe covered by the data.

## 3.0 Data Preparation 

### 3.1 Importing libraries

This cell imports essential libraries for our customer churn analysis. 

```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

### 3.2 Loading and viewing the Dataset

We then load the customer churn analysis data from a CSV file, and perform data exploration to understand our data.

```python
def analyze_data(file_path):

# Load the dataset
    df = pd.read_csv(file_path)
    
# Display the first few rows of the DataFrame
    print(df.head())

# Display the last few rows of the DataFrame
    print(df.tail())

# Describe the data
    print(df.describe().transpose())

# Provide info about the data
    print(df.info())

# Check for missing values
    print(df.isnull().sum())

# Visualize missing data
    msno.bar(df)

# Check for duplicate rows
    duplicates = df.duplicated()
    print(f"Number of duplicate rows = {duplicates.sum()}")

    return df

df = analyze_data('Churn_Telecom_Dataset.csv')
```

### 3.3 Viewing Unique Values

It is always important to view unique values in a dataset since this stage helps you identify potential data quality issues. These issues could include inconsistent formatting of categorical data (e.g., "CA" and "California" for the same state) which can lead to errors in analysis. Examining unique values can reveal these inconsistencies.

```python
# Function to inspect each column for unique values

def inspect_columns(df):
    """Prints unique values of each column in the DataFrame."""
    for col in df.columns:
        print(f"{col}: \n{df[col].unique()}\n")

inspect_columns(df)
```

Data Types: The dataset contains a variety of data types. We have integer values (such as 'account length', 'area code', 'number vmail messages'), floating-point numbers (like 'total day minutes', 'total day charge'), and string or object types (including 'state', 'phone number', and others). We will use the different types differently.

Missing Values: We can see that our dataset has no missing values. This makes our work easier. 

We also notice from the unique values that our data is consistent. 

Excluding Some Columns: Though the 'area code' and 'phone number' columns provide valueable contact information of our customers, we will drop the 'area code' column to prevent overfitting of our model (in case there are other area codes that haven't been included in our data), and drop the 'phone number' column to maintain our customers' privacy.

### 3.4 Dropping Unnecessary Columns

We will drop columns that won't necessarily help us in our modeling and prediction. 

```python
def drop_columns(df):

    df = df.drop(['area code', 'phone number'], axis=1)
    return df

# Drop columns
df = drop_columns(df)

df
```

### 3.5 Checking for Outliers

We will check for outliers in our dataset, then make a decision whether to remove them or not. 

```python
# Create box plots for specified columns

def create_box_plots(df):
    columns = ['account length', 'total day minutes', 'total day calls', 'total day charge', 
               'total eve minutes', 'total eve calls', 'total eve charge', 
               'total night minutes', 'total night calls', 'total night charge', 
               'total intl minutes', 'total intl calls', 'total intl charge']
    
# Calculate the number of rows needed
    n_columns = len(columns)
    n_rows = (n_columns // 4) + (1 if n_columns % 4 != 0 else 0)
    
    fig, axs = plt.subplots(n_rows, 4, figsize=(15, 5 * n_rows))
    
# Flatten the array to make indexing easier
    axs = axs.flatten()
    
    for i, column in enumerate(columns):
        if column in df.columns:
            sns.boxplot(y=df[column], ax=axs[i])
            axs[i].set_title(column)
            
        else:
            print(f"Warning: Column '{column}' not found in DataFrame")

# Remove any empty subplots
    for j in range(i + 1, len(axs)):
        fig.delaxes(axs[j])
    
    plt.tight_layout()
    plt.show()

create_box_plots(df)
```

While the boxplots reveal outliers, I've opted to keep them in the data. This will allow our models to be exposed to a wider range of data points, enhancing their ability to generalize to unseen scenarios during prediction.

## 4.0 Exploratory Data Analysis

### 4.1 Univariate Data Analysis

#### 4.1.0 Check for Counts for Columns with Numeric Features

We will create bar charts to understand the distribution of values within each numeric feature, which is crucial for identifying any skewness, outliers, or unusual patterns.

```python
# List of numeric features
numeric_features = df.select_dtypes(include=['int64', 'float64']).columns
print("Numeric Features:", numeric_features)

# Count unique values for each numeric feature
unique_counts = df[numeric_features].nunique()
print("Unique Value Counts for Numeric Features:\n", unique_counts)

# Plot histograms
df[numeric_features].hist(bins=30, figsize=(15, 10))
plt.suptitle('Histograms of Numeric Features', y=1.02)
plt.show()
```

Features like total day minutes, total day charge, total eve minutes, total eve charge, total night minutes, total night charge, total intl minutes, total intl charge have a large number of unique values, indicating that they are likely continuous variables.

Additionally, features total day minutes, total day charge, total eve minutes, total eve charge, total night minutes, total night charge, total intl minutes, and total intl charge seem to be related. The charges are likely derived directly from the corresponding minutes, implying a strong linear relationship.

Continuous features like total day minutes, total eve minutes, total night minutes, total intl minutes, and their corresponding charges will need to be scaled before modeling.

#### 4.1.1 Check for Counts for Rows with Categorical Features

```python
# List of categorical features
categorical_features = df.select_dtypes(include=['object', 'category']).columns
print("Categorical Features:", categorical_features)

# Count unique values for each categorical feature
unique_value_counts = df[categorical_features].nunique()
print("Unique Value Counts for Categorical Features:\n", unique_value_counts)

# Display frequency counts for each categorical feature
for feature in categorical_features:
    print(f"\nFrequency count for {feature}:")
    print(df[feature].value_counts())

# Plot bar charts for each categorical feature
plt.figure(figsize=(15, 10))
for i, feature in enumerate(categorical_features):
    plt.subplot(len(categorical_features), 1, i+1)
    sns.countplot(y=df[feature], order=df[feature].value_counts().index)
    plt.title(f'Distribution of {feature}')
    plt.xlabel('Count')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```
